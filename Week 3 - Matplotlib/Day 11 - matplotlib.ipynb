```python
import matplotlib.pyplot as plt # importing matplotlib
import numpy as np # importing numpy

'''--------------------
Generating a graph (plot)
---------------------
It's a 3-stage process
1. Generate a Figure object. (Contains all the plot elements)
2. Generate an Axes object. ( (subplots) the region of the image with the data space. A given figure can contain many Axes )
3. Generate a plot. 
'''
y = np.linspace(-1,1,5) # returns an evenly spaced numbers over a specified interval
x = np.arange(5) #returns and array of evenly spaced values. 
y2 = np.linspace(-2,2,5)

print(x)
print(y) 
print(y2)
```

```python
fig = plt.figure()
ax = fig.subplots()


line1 = ax.plot(x,y) # the plot() function is used to draw points (markers) in a diagram. For this take 
# the array 'x' and us it for the horizontal position then take the array 'y' and use it for the virtical position
line2 = ax.plot(x,y2) 

print(fig)
print(ax)
print(line1)
```

```python
# 6 subplots in a 3x2 grid

n_rows = 3
n_columns = 2
fig2 = plt.figure()
ax2 = fig2.subplots(n_rows,n_columns) # creating multiple gragh spaces
ax2[0,1].plot(x,y) # ploting a line in the graph space at location row 0 column 1 
```

```python
#basic plotting functions
y1 = np.arange(0,110,10) # numpy arrange creates an array from 0 to 110 at 10 increments. (not including 110)
y2 = np.random.random(11) # creates an array of 11 elements of random floats
x = np.arange(11) # creates an array of integers starting from 0 to 1 before 11 so 0-10 

# fig = plt.figure()  ax = fig.subplots()
fig, ax = plt.subplots(2,2) #   not sure what this is doing but the graph doesnt show without it.
ax[0,0].plot(x,y1) # default line chart
ax[0,1].scatter(x,y2) # scatter graph
ax[1,0].bar(x,y1) # bar chart
ax[1,1].barh(x,y1) # horizontal bar chart

print(y1)
print(y2)
print(x)
```

```python
# Trigonometric functions
x = np.arange(0, 4*np.pi, 0.05)
ycos = np.cos(x)
ysin = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, ycos, 'b-', label="cos(x)") # 'b-' produces a solid blue line
ax.plot(x, ysin, 'r--', label="sin(x)") # 'r--' produces a dashed red line.

ax.set_title("Trigonometric functions")
ax.set_xlabel("0 to 4pi")
ax.set_ylabel("0 to 4pi")
ax.legend()

```

```python
# Bar plot 
people = ["Student A", "Student B"]
studentA = np.array([90,50,80,40])
studentB = np.array([75,45,60,95])
x = np.arange(len(studentA))

fig, ax = plt.subplots(1,2)

ax[0].bar(x, studentA, width=0.3)
ax[1].bar(x, studentB, width=0.3)
```
