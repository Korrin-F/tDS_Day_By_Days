```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw a line

x_points = np.array([1,8]) # these will be the x (horizontal) points
y_points = np.array([3,10]) # these will be the y (virtical) points

plt.plot(x_points,y_points)
```

```python
# Draw points instead of a connected line
y2_points = np.array([4,9])
plt.plot(x_points,y_points, 'o') #shortcut string notation 'o' creates dots on the graph
plt.plot(x_points, y2_points, '*') 
```

```python
# multiple points

x3_points = np.array([1,2,6,8])
y3_points = np.array([3,8,1,10])

plt.plot(x3_points,y3_points, 'o-')
```

```python
# Default x-points

y4_points = np.array([3,8,1,10,5,7])

plt.plot(y4_points)
```
