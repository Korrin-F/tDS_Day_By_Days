```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Markers
# colors are auto-generated
circle = plt.plot(1,1,'o')
star = plt.plot(2,2,'*')
point = plt.plot(3,3,'.')
pixel = plt.plot(4,4,',')
x = plt.plot(5,5,'x')
x_filled = plt.plot(6,6,'X')
plus = plt.plot(7,7,'+')
plus_filled = plt.plot(8,8,'P')
square = plt.plot(9,9,'s')
diamond = plt.plot(10,10,'D')
diamond_thin = plt.plot(11,11,'d')
pentagon = plt.plot(12,12,'p')
hexagon = plt.plot(13,13,'H')
hexagon_same = plt.plot(14,14,'h')
triangle_down = plt.plot(15,15,'v')
triangle_up = plt.plot(16,16,'^')
triangle_left = plt.plot(17,17,'<')
triangle_right = plt.plot(18,18,'>')
tri_down = plt.plot(19,19,'1')
tri_up = plt.plot(20,20,'2')
tri_left = plt.plot(21,21,'3')
tri_right = plt.plot(22,22,'4')
virtical_line = plt.plot(23,23,'|')
horizontal_line = plt.plot(24,24,'-')

```

```python
# line references 
# colours automatically generate
solid_y = np.random.randint(0,100,5) # 5 random integers from 0 to 100 
solid = plt.plot(solid_y, '-') # a solid line is default and does not need to be specified

dotted_y = np.random.randint(0,100,5)
dotted = plt.plot(dotted_y, ':')

dashed_y = np.random.randint(0,100,5)
dashed = plt.plot(dashed_y, '--')

dash_dot_y = np.random.randint(0,100,5)
dash_dot = plt.plot(dash_dot_y, '-.')
```

```python
# color reference
# the short color value can be one of the following:

red_y = np.random.randint(0,100,5)
red = plt.plot(red_y, 'r')

green_y = np.random.randint(0,100,5)
green = plt.plot(green_y, 'g')

blue_y = np.random.randint(0,100,5)
blue = plt.plot(blue_y, 'b')

cyan_y = np.random.randint(0,100,5)
cyan = plt.plot(cyan_y, 'c')

magenta_y = np.random.randint(0,100,5)
magenta = plt.plot(magenta_y, 'm')

yellow_y = np.random.randint(0,100,5)
yellow = plt.plot(yellow_y, 'y')

black_y = np.random.randint(0,100,5)
black = plt.plot(black_y, 'k')

white_y = np.random.randint(0,100,5)
white = plt.plot(white_y, 'w')
```

```python
# shortcut string parameter 
# use this to combine the marker, line and color shortcuts together
#  marker/line/color 'o:r'

diamond_dash_black_y = np.random.randint(0,100,5)
diamond_dash_black = plt.plot(diamond_dash_black_y, 'd--k')

circle_dot_cyan_y = np.random.randint(0,100,5)
circle_dot_cyan = plt.plot(circle_dot_cyan_y, 'o:c')

x_solid_magenta_y = np.random.randint(0,100,5)
x_solid_magenta = plt.plot(x_solid_magenta_y, 'x-m')

```

```python
# marker size
# ms = marker size
m_size_y = np.random.randint(0,100,5)
m_size = plt.plot(m_size_y, 'X-b', ms = 20)
```

```python
# marker colours expanded 

# marker edge colour (mec)
edge_colour_red_y = np.random.randint(0,100,5)
edge_colour_red = plt.plot(edge_colour_red_y, 'o-g', ms = 20, mec = 'r')

# marker face colour (mfc)
face_colour_cyan_y = np.random.randint(0,100,5)
face_colour_cyan = plt.plot(face_colour_cyan_y, 'o:m', ms = 20, mfc = 'c') 

# combining both the face and edge 
face_edge_yellow_y = np.random.randint(0,100,5)
face_edge_yellow = plt.plot(face_edge_yellow_y, 'o--b', ms = 20, mfc = 'y', mec = 'y')

# using hex values
hex_colours_y = np.random.randint(0,100,5)
hex_colours = plt.plot(hex_colours_y, '#bc436b', marker = '*', mec = '#ff8277', mfc = '#ff8277', ms = 20)

# using colour names
colour_names_y = np.random.randint(0,100,5)
colour_names = plt.plot(colour_names_y, 'orange', marker = 's', mec = 'hotpink', mfc = 'hotpink', ms = 20)
```
