```python
#--------------------------
# initialising a numpy array .zeros , .ones , .full
#--------------------------

number_of_rows = 2 # store a value for the number of rows you want in your array
number_of_columns = 2 # store a value for the number of columns you want in your array 

# only zeros and ones are built in functions
# for other numbers see .full function below
arrays_of_zeros = np.zeros((number_of_rows,number_of_columns)) # returns a new array of given shape and type filled with zeros
arrays_of_ones = np.ones((number_of_rows,number_of_columns)) # returns a new array of given shape and type filled with ones
print(f"An array of zeros with {number_of_rows} rows & {number_of_columns} columns(elements): ")
print(arrays_of_zeros)
print(f"An array of ones with {number_of_rows} rows & {number_of_columns} columns(elements): ")
print(arrays_of_ones)

# Use .full to return a new array of given shape and type filled with a specified number
arrays_of_5s = np.full((number_of_rows, number_of_columns), 7)
print(f"An array of fives with {number_of_rows} rows & {number_of_columns} columns(elements): ")
print(arrays_of_5s)

# perform arithmatic on the elements to change the number again
five_times_seven = 5 * arrays_of_5s
print(f"An array of fives * sevens with {number_of_rows} rows & {number_of_columns} columns(elements): ")
print(five_times_seven)
```

```python
#--------
# Task
#--------

# Take this array and identifiy all the odd numbers then replace them with thei negative counterparts
odd_arr = np.array([0,2,4,5,3,7,8,9,5,1])

# MY SOLUTION
for index, i in enumerate(odd_arr) :
    if (i % 2 == 1):
        i = i * -1
        odd_arr[index] = i
        #print(i)

print("My solution: ", odd_arr)

odd_arr2 = np.array([0,2,4,5,3,7,8,9,5,1])

# ALT SOLUTION
odd_arr2[odd_arr2 % 2 ==1] = -1 * odd_arr2[odd_arr2 % 2 ==1] 
# â˜ðŸ» left side returns a boolean value in the [] brackets then pulls that figure from the array.
# the right side multiplies that same value by minus 1 to change it to a negative value
print("Left side in [ ] test: ", odd_arr2 % 2 ==1)
print("Full left side test: ", odd_arr2[odd_arr2 % 2 ==1])
print("Right side test: ", -1 * odd_arr2[odd_arr2 % 2 ==1] ) 

print("Class soultion: ", odd_arr2)
```

```python
# NumPy array arithmetics (slide 13)

a = np.array([[1,2,5],[7,1,3]])
b = np.array([[4,6,9],[2,5,1]])

# addition 
add1 = a + b # top number plus the bottom number
print(add1)

add2 = np.add(a,b)
print(add2) #performs the same function as add1

# subtraction
sub1 = a - b
print(sub1)

sub2 = np.subtract(a,b) # behaves the same as sub1
print(sub2)

# multiplication
mul1 = a * b
print(mul1)

mul2 = np.multiply(a,b) #behaves the same as mul1
print(mul2)

# division
div1 = a/b 
print(div1)

div2 = np.divide(a,b) #behaves the same as mul1
print(div2)

# Numpi array can do much more...
#exponential
#square root
#sin
#cos
#log
```

```python
# Maximum overall number  (slide 14)
maximum = a.max() # returns the largest value in the list
print(maximum)

# standard deviation
standard_dev = np.std(a)
print(standard_dev)
```

```python
#-----------------------
# NumPy array manipulation
#-----------------------

b = np.array([[1,2,3],[4,5,6]])
print(b)

#-----------------------
# Transpose .transpose
#----------------------
# swaps the number of rows with the number of columns

b_trans = np.transpose(b)
print(b_trans)

#-----------------------
#Reshape .reshape
#----------------------
reshape_b = np.reshape(b, (3,2))
print(reshape_b)


#---------------------
# Stacking .vstack .hstack
#---------------------
print("This is list a: ")
print(a)
print("This is list b: ")
print(b)

# horizontal
a_b_horiz = np.hstack([a,b])
print("The first array in a is in the same row as the first array in b. A new row is made for further arrays: ")
print(a_b_horiz)

# virtical
a_b_virt = np.vstack([a,b])
print("The arrays from a are stacked on top of the arrays from b.: ")
print(a_b_virt)
```

```python
#-----------------
# Concacenate and array .concatenate
#-----------------

# concatenate by the row using 0
a_b_row_conc = np.concatenate((a,b), axis=0)
print(a_b_row_conc)

# concatenate by the column using 1
a_b_col_conc = np.concatenate((a,b), axis=1)
print(a_b_col_conc)
```
