```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import LinearRegression
```

```python
#--------------------------------------------------
# tuning lasso ridge.py
#--------------------------------------------------

data = pd.read_csv('uni_data.csv')  # load our data

X = data.iloc[:-1, 1:9] # set predicting variables
y = data.iloc[:-1, 9] # set variable to predict (score)

# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

# define model
model = RidgeCV(alphas=np.arange(0.5, 1, 0.01), cv=cv, scoring='neg_mean_absolute_error')

# fit model
model.fit(X, y)

# summarize chosen configuration
print('alpha: %f' % model.alpha_)



```

```python
#--------------------------------------------------
# ols_linear_regression.py
#--------------------------------------------------

data = pd.read_csv('car_data.csv')  # load our data

# select the 2nd and 3rd columns and convert them to a numpy array
X = data.iloc[:, 2].values.reshape(-1, 1)
Y = data.iloc[:, 3].values.reshape(-1, 1)

linear_regressor = LinearRegression()  # create a linear regression object
linear_regressor.fit(X, Y)  # perform the linear regression
Y_pred = linear_regressor.predict(X)  # make the predictions based on the model

# plotting data and predicted variable
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.xlabel('Selling price')
plt.ylabel('Km driven') 
plt.show()
```
